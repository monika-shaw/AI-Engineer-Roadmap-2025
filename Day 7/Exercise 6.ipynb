```python
Given a CSV file of transactions, show:

The total spent per customer.

List customers who spent more than 250.

Save summarized results to a new CSV.
```

```python
import pandas as pd

df = pd.read_csv("transactions.csv")

summary = df.groupby("customer")["amount"].sum().reset_index()

big_spenders = summary[summary["amount"] > 250]

summary.to_csv("transactions_summary.csv", index=False)
big_spenders.to_csv("big_spenders.csv", index=False)

print("Customers who spent more than 250:")
print(big_spenders)
```
