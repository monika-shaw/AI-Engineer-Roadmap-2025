```python
Given a CSV file containing student grades, perform the following:

Load the CSV into a DataFrame.

Fill missing values.

Calculate average score for each student.

Find the top scorer by average.

Save cleaned data to a new CSV file.
```

```python
import pandas as pd

df = pd.read_csv("students_grade.csv")

df_filled = df.fillna(df.mean(numeric_only=True))

df_filled['average'] = df_filled[['math', 'science', 'english']].mean(axis=1)


top_student = df_filled.loc[df_filled['average'].idxmax(), 'student']
top_score = df_filled['average'].max()
print(f"Top scorer: {top_student} with average {top_score:.2f}")


df_filled.to_csv("students_grades_cleaned.csv", index=False)
```
