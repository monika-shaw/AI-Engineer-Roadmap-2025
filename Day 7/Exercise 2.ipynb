```python
Analyze student scores to find average, highest, lowest marks per subject and plot the distribution.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('students.csv')
print(df)
```

```python
average_score = df.mean(numeric_only =True)
print(average_score)
```

```python
max_score = df.max(numeric_only = True)
print(max_score)
```

```python
min_score = df.min(numeric_only = True)
print(min_score)
```

```python
plt.hist(df[['Math','Science','English']])
plt.title("Score Distribution")
plt.xlabel("Marks")
plt.show()
```
