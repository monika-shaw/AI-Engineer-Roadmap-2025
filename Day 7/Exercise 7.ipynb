```python
Given two CSV files (employees and salaries), perform the following:

Merge the datasets on employee ID

Calculate salary increases from previous year

Create a pivot table showing average salary by department and year

Handle missing data appropriately
```

```python
import pandas as pd

employees = pd.read_csv("employees.csv")
salaries = pd.read_csv("salaries.csv")
```

```python
employees['hire_date'] = pd.to_datetime(employees['hire_date'])

merged_df = pd.merge(employees, salaries, on='emp_id', how='inner')
```

```python
dept_avg = merged_df.groupby('department')['salary'].transform('mean')
merged_df['salary'] = merged_df['salary'].fillna(dept_avg)
```

```python
salary_pivot = merged_df.pivot_table(
    values='salary', 
    index='emp_id', 
    columns='year', 
    fill_value=0
)
```

```python
salary_pivot['increase_pct'] = ((salary_pivot[2024] - salary_pivot[2023]) / salary_pivot[2023] * 100).round(2)
```

```python
summary = merged_df.pivot_table(
    values='salary',
    index='department',
    columns='year',
    aggfunc='mean',
    fill_value=0
).round(2)
```

```python
salary_pivot.to_csv("salary_analysis.csv")
summary.to_csv("department_summary.csv")

print("Salary increases by employee:")
print(salary_pivot[['increase_pct']])
print("\nAverage salary by department and year:")
print(summary)
```
