```python
Q1. Compute cosine similarity between two vectors
```

```python
import numpy as np

# Two feature vectors
a = np.array([2, 3, 4])
b = np.array([1, 0, 6])

# Cosine similarity
cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
print("Cosine Similarity:", cos_sim)
```

```python
Q2. Implement matrix multiplication for neural network layer
```

```python
import numpy as np

X = np.array([[1, 2, 3]])
W = np.array([[0.2, 0.8],
              [0.5, 0.1],
              [0.9, 0.4]])

y = np.dot(X, W)
print("Layer Output:", y)
```

```python
Q3. Compute optimal weights using Normal Equation
```

```python
import numpy as np

X = np.array([[1, 1],
              [1, 2],
              [1, 3],
              [1, 4]])   # include bias (1s)
y = np.array([[2],
              [3],
              [4],
              [5]])

theta = np.linalg.inv(X.T @ X) @ X.T @ y
print("Optimal Weights (theta):", theta)
```

```python
Q4. Perform PCA using Eigen Decomposition
```

```python
import numpy as np

X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0]])

X_mean = X - np.mean(X, axis=0)

cov_mat = np.cov(X_mean, rowvar=False)

eigen_values, eigen_vectors = np.linalg.eig(cov_mat)

sorted_idx = np.argsort(eigen_values)[::-1]
eigen_vectors = eigen_vectors[:, sorted_idx]

X_pca = np.dot(X_mean, eigen_vectors[:, :1])
print("PCA Reduced Data:\n", X_pca)
```

```python
Q5. Implement simple gradient descent
```

```python
import numpy as np

x = 0.0  
lr = 0.1 

for i in range(10):
    grad = 2*(x - 3) 
    x = x - lr * grad
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {(x-3)**2:.4f}")
```
