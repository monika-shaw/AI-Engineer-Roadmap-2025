```python
Q1. Compute cosine similarity between two vectors
```

```python
import numpy as np

# Two feature vectors
a = np.array([2, 3, 4])
b = np.array([1, 0, 6])

# Cosine similarity
cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
print("Cosine Similarity:", cos_sim)
```

```python
Q2. Implement matrix multiplication for neural network layer
```

```python
import numpy as np

X = np.array([[1, 2, 3]])
W = np.array([[0.2, 0.8],
              [0.5, 0.1],
              [0.9, 0.4]])

y = np.dot(X, W)
print("Layer Output:", y)
```

```python
Q3. Compute optimal weights using Normal Equation
```

```python
import numpy as np

X = np.array([[1, 1],
              [1, 2],
              [1, 3],
              [1, 4]])   # include bias (1s)
y = np.array([[2],
              [3],
              [4],
              [5]])

theta = np.linalg.inv(X.T @ X) @ X.T @ y
print("Optimal Weights (theta):", theta)
```
