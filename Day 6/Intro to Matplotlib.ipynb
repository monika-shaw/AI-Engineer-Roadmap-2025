```python
import matplotlib.pyplot as plt
print(matplotlib.__version__)
x =[1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.title("Grpah")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
plt.plot([1,2,3,4], [2,4,6,8], color='red', linestyle='--')
plt.title("Dashed Red Line")
plt.show()
```

```python
plt.plot([1,2,3],[3,2,5])
plt.title("With Grid")
plt.grid(True)
plt.show()
```

```python
fruits = ['Apple', 'Banana', 'Mango', 'Grapes']
sales = [30, 20, 40, 25]
plt.bar(fruits, sales, color='orange')
plt.title("Fruit Sales")
plt.xlabel("Fruit")
plt.ylabel("Quantity")
plt.show()
```

```python
import numpy as np
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x, y, color='purple')
plt.title("Random Scatter Plot")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1,2,3],[1,4,9])
plt.title("Left Plot")

plt.subplot(1, 2, 2)
plt.plot([1,2,3],[1,2,3])
plt.title("Right Plot")

plt.show()
```
