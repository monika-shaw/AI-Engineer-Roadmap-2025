```python
# Day 5: Calculus Project â€” Gradient Descent from Scratch

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 5*x + 6

def df(x):
    return 2*x + 5

x_old = 5 
learning_rate = 0.1
epochs = 30

x_values = [x_old]
y_values = [f(x_old)]

for i in range(epochs):
    gradient = df(x_old)
    x_new = x_old - learning_rate * gradient
    x_old = x_new
    
    x_values.append(x_new)
    y_values.append(f(x_new))
    
    print(f"Iteration {i+1}: x = {x_new:.4f}, f(x) = {f(x_new):.4f}")


```

```python
x = np.linspace(-10, 5, 100)
y = f(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x) = xÂ² + 5x + 6')
plt.scatter(x_values, y_values, color='red')
plt.plot(x_values, y_values, color='red', linestyle='--', label='Gradient Descent Path')
plt.title("Gradient Descent Using Calculus Derivative")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
