```python
1. Derivative of a Function
```

```python
import sympy as sp

x = sp.Symbol('x')
f = 3*x**3 - 5*x**2 + 7*x - 2

f_prime = sp.diff(f, x)
f_prime
```

```python
2. Find Critical Points (Optimization)
```

```python
f = x**3 - 6*x**2 + 9*x + 1

f_prime = sp.diff(f, x)

critical_points = sp.solve(f_prime, x)
critical_points
```

```python
3. Compute Definite Integral
```

```python
f = x**2 + 2*x + 1

area = sp.integrate(f, (x, 0, 3))
area
```
